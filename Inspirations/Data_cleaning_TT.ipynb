```python
# process all transcripts at once
transcripts = {}
for filename in os.listdir("maxqda_export/Liverpool"):
    if filename.endswith(".txt"):
        path = f"maxqda_export/Liverpool/{filename}"
        transcripts[filename] = parse_transcript(path)
    if filename.endswith(".xlsx"):
        path = f"maxqda_export/Liverpool/{filename}"
        transcripts[filename] = parse_transcript_xlsx(path)

# access individual ones like
#transcripts["Session10_Transcript.txt"]

# or combine all into one big df with a session column
df_all = pl.concat([
    df.with_columns(pl.lit(name).alias("session"))
    for name, df in transcripts.items()
], how="vertical_relaxed")

transcripts.keys()
```
